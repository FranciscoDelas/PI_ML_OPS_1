```python
import pandas as pd
import json
import ast
from tabulate import tabulate
```

## Funcion 01
#### def developer( desarrollador : str ): Cantidad de items y porcentaje de contenido Free por año según empresa desarrolladora. Ejemplo de retorno:
Año	Cantidad de Items	Contenido Free
2023	50	27%
2022	45	25%
xxxx	xx	xx%

```python
row2 = []
with open('Datasets/steam_games.json', 'r') as f:
        for line in f.readlines():
            data = json.loads(line)
            row2.append(data)

dfgame = pd.DataFrame(row2)
dfgame.dropna(how='all', inplace=True)
dfgame.drop_duplicates(subset='id', keep='first', inplace=True)
dfgame.head()
```

```python
dfgame.info()
```

```python
import pandas as pd

# Mantén solo las columnas necesarias
dfgame = dfgame[['release_date', 'developer', 'price']]

# Convierte la columna 'release_date' al formato de fecha
dfgame['release_date'] = pd.to_datetime(dfgame['release_date'], errors='coerce')

# Elimina filas con valores faltantes en las columnas relevantes
dfgame.dropna(subset=['release_date', 'developer', 'price'], inplace=True)

# Agrega una nueva columna 'year' que contenga solo el año de 'release_date'
dfgame['year'] = dfgame['release_date'].dt.year

# Mantén solo las columnas necesarias para el cálculo
dfgame = dfgame[['year', 'developer', 'price']]


# Ahora 'dfgame' contiene solo las columnas necesarias y los valores necesarios de formato correcto.
# Podemos utilizar este DataFrame para realizar la función 'developer'.
```

```python


def developer(desarrollador):
    # Filtra el DataFrame para obtener los datos del desarrollador especificado
    dfgame_filtered = dfgame[dfgame['developer'] == desarrollador]

    # Agrupa por año y calcula la cantidad de ítems y el porcentaje de contenido gratuito
    result = dfgame_filtered.groupby('year').agg(
        Cantidad_de_Items=('developer', 'count'),
        Contenido_Free_porcentual =('price', lambda x: (x.astype(str).str.contains('Free', case=False, regex=False).sum()) / len(x) * 100)
).reset_index()

    # Imprime la tabla de resultados
    print(tabulate(result, headers='keys', tablefmt='psql'))

# Llama a la función con el nombre del desarrollador como argumento
nombre_desarrollador = 'Ankama Studio'  # Reemplaza 'NombreDelDesarrollador' con el nombre real del desarrollador
developer(nombre_desarrollador)
```
