```python
import pandas as pd
import json
import ast
```

Creamos el DF REVIEWS

```python
#interpretamos cada línea como una expresión literal de Python utilizando ast.literal_eval(), y luego crea un DataFrame de Pandas

rows2 = []
with open('Datasets/user_reviews.json', 'rt',encoding= 'utf-8') as f:
    for line in f.readlines():
        rows2.append(ast.literal_eval(line))
dfreviews = pd.DataFrame(rows2)
dfreviews.head()
```

```python
dfreviews.info()
```

Limpieza

```python
dfreviews.dropna(how='all', inplace=True)
```

Desanidamos columna 'reviews' y tomamos subcolumnas como nuevas columnas del dfreviews

```python
dfreviews = dfreviews.explode('reviews')
dfreviews
```

```python
dfreviews['item_id'] = dfreviews['reviews'].apply(lambda x: x.get('item_id') if isinstance(x, dict) else '')
dfreviews['review'] = dfreviews['reviews'].apply(lambda x: x.get('review') if isinstance(x, dict) else '')
```

```python
dfreviews.head()
```

```python
dfreviews=dfreviews[['user_id','item_id','review']]
dfreviews.dropna(subset='review', inplace=True)
```

```python
dfreviews.head()
```

```python
dfreviews.info()
```

ANALISIS DE SENTIMIENTO sobre columna 'review'

```python
from textblob import TextBlob
# Función para realizar el análisis de sentimiento
def sentiment_analysis(text):
    if isinstance(text, str):
        analysis = TextBlob(text)
        polarity = analysis.sentiment.polarity
        if polarity < 0:
            return 0  # Sentimiento negativo
        elif polarity == 0:
            return 1  # Sentimiento neutral
        else:
            return 2  # Sentimiento positivo
    else:
        return 1  # Valor por defecto para valores nulos o no strings

# Aplicar la función a la columna 'review' y crear la nueva columna 'sentiment_analysis'
dfreviews['sentiment_analysis'] = dfreviews['review'].apply(sentiment_analysis)
```

```python
dfreviews
```

```python
dfreviews=dfreviews[['user_id','item_id','sentiment_analysis']]
dfreviews.head()
```
